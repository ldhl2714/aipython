```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic') #폰트설정
df = pd.DataFrame({'도시':['서울','부산','인천','대구'], '인구 수':[9904312,3448737,2890451,2466052]})
plt.bar(df['도시'],df['인구 수'])
plt.xlabel('도시')
plt.ylabel('인구 수')
plt.show()
```

```python
import seaborn as sns
iris_data=pd.read_csv('/iris.csv')
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data= iris_data)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
data = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})
sns.scatterplot(x='x',y='y',hue='category',data=data)
plt.title('Scatter Plot by seaborn')
plt.show()
```

```python
data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})
sns.countplot(x='category',data=data)
plt.show()
```

```python
data=pd.read_csv('/penguins.csv')
data.info()
sns.barplot(x='species',y='body_mass_g',data=data,hue='sex',errorbar='sd')
plt.title('펭귄 품종별 몸무게의 평균')
plt.show()

sns.pairplot(data,hue='species')
plt.title('항목들의 상관관계')
plt.show()

sns.regplot(x='flipper_length_mm',y='body_mass_g', data=data)
plt.title('산점도와 추세선')
plt.show()
```
