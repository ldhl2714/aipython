```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris_data=pd.read_csv('/iris.csv')
iris_data.info()
iris_data. describe()
class_count=iris_data.groupby('species').count()
print(class_count)
```

```python
iris_data.isnull().sum() # 결측치 확인
iris_data[iris_data.duplicated()] #중복값 확인
data=iris_data.drop_duplicates() #중복값 제거
data[data.duplicated()] #제거 확인
data.info()
```

```python
sum_spe=data.groupby('species').sum()
mean_spe=data.groupby('species').mean()
print('합계',sum_spe)
print('평균',mean_spe)
```

```python
plt.bar(data['species'], data['sepal_length'])
plt.show()
sns.barplot(x='species',y='sepal_length',data=data)
plt.show()
```

```python
data, axs = plt.subplots(2,2)

axs[0,0].hist('sepal_length')
axs[0,1].hist('sepal_width')
axs[1,0].hist('petal_length')
axs[1,1].hist('petal_width')
plt.show()
```

```python
data.hist('sepal_length')
plt.show()
data.hist('sepal_width')
plt.show()
data.hist('petal_length')
plt.show()
data.hist('petal_width')
plt.show()
```
