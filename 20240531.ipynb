#머신러닝

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model #scikit-learn 모듈을 가져옴.
regr = linear_model.LinearRegression() #선형회귀 모델 생성

X = [[164], [179], [162], [170]] #키 데이터
y = [53, 63, 55, 59] #몸무게 데이터

regr.fit(X, y) #회귀
score = regr.score(X, y) #회귀평가(예측을 잘하는지)
print("The score of this line for the data: ", score)

input_data = [[180], [185]] #새로 예측해 볼 데이터
result = regr.predict(input_data)
print(result)

plt.scatter(X, y)
y_pred = regr.predict(X)
plt.plot(X, y_pred, color = 'blue', linewidth = 3)
plt.show()
```

#Student_Marks

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv('Student_Marks.csv')
df.describe()
df.info()
df.head()
df.tail()
```

```python
df.corr()
fig, ax=plt.subplots(ncols=2, figsize=(12,6))
sns.scatterplot(data=df, x='number_courses',y='Marks', ax=ax[0])
sns.scatterplot(data=df, x='time_study',y='Marks', ax=ax[1])
plt.show()
```

```python
#데이터셋 분할(훈련-테스트)
from sklearn.model_selection import train_test_split
x=df.drop('Marks',axis=1)
y=df['Marks']
X_train, X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X_train,Y_train)
```

```python
fig, ax=plt.subplots(ncols=2, figsize=(12,6))
sns.regplot(x=df['time_study'],y=df['Marks'], ax=ax[0])
sns.regplot(x=df['number_courses'],y=df['Marks'], ax=ax[1])
plt.show()
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr_pred=regr.predict(X_test)
r2_score(regr, Y_test)
mean_squared_error(regr,Y_test)
```
